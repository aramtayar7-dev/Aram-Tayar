```python
# سەرەتا ژمارەی ڕۆژەکان بە سفر دەست پێدەکات
days_clean = 0

def recovery_guide_bot(user_input):
    global days_clean # ئەمە بۆ ئەوەیە بەرنامەکە ژمارەی ڕۆژەکان لەبیر نەکات

    text = user_input.lower()

    # لیستی وشەکان
    zianekan = ["زیان", "خراپە", "بۆچی خراپە", "کێشە", "مەترسی"]
    charasar = ["چارەسەر", "چۆن واز بهێنم", "وازهێنان", "یارمەتی", "ڕێگا", "سروشتی"]
    arezakan = ["ئارەزوو", "خەریکە", "بێزارم", "بێهێزم", "تێکچووم"]

    # لیستی وشەکانی تایبەت بە ژمێرەری ڕۆژەکان
    ziadkrdn = ["سەرکەوتم", "ڕۆژێکی تر", "تێپەڕاند"]
    shkandin = ["شکاندم", "گەڕامەوە", "سفر", "هەڵەم کرد"]
    prsiari_roj = ["چەند ڕۆژە", "ڕۆژەکانم", "چەندم"]

    # مەرجەکانی ژمێرەری ڕۆژەکان
    if any(word in text for word in ziadkrdn):
        days_clean += 1
        return f"🎉 پیرۆزە! تۆ زۆر بەهێزیت. ئەمڕۆش بە سەرکەوتوویی تێپەڕی.\nئێستا ژمارەی ڕۆژە پاکەکانت بوو بە: {days_clean} ڕۆژ."

    elif any(word in text for word in shkandin):
        days_clean = 0
        return "هیچ کێشە نییە، هەموومان هەڵە دەکەین. گرنگ ئەوەیە کۆڵ نەدەیت و لێوەی فێربیت. ژمێرەرەکە سفر کرایەوە، با بەیەکەوە سەرلەنوێ دەست پێبکەینەوە! تۆ دەتوانیت."

    elif any(word in text for word in prsiari_roj):
        if days_clean == 0:
            return "ئێستا لە ڕۆژی سفرداین. با هەر ئەمڕۆ بکەینە یەکەم ڕۆژی دەستپێکردنێکی نوێ!"
        else:
            return f"تۆ تا ئێستا {days_clean} ڕۆژە سەرکەوتوو بوویت. بەردەوام بە، زۆرت نەماوە مێشکت بە تەواوی ئاسایی ببێتەوە!"

    # مەرجەکانی پێشوو
    elif any(word in text for word in zianekan):
        return """سەیرکردنی ئەم جۆرە ڤیدیۆیانە چەندین زیانی هەیە، لەوانە:
١. تێکدانی سیستەمی دۆپامین لە مێشکدا.
٢. کەمبوونەوەی سەرنج و تەرکیز.
٣. دروستبوونی دڵەڕاوکێ و خەمۆکی.
٤. تێکدانی پەیوەندییە کۆمەڵایەتییەکان."""

    elif any(word in text for word in charasar):
        return """بۆ وازهێنان، ئەم هەنگاوانە پەیڕەو بکە:
١. کاتە بەتاڵەکانت پڕ بکەرەوە.
٢. مۆبایل مەبە بۆ ناو جێگەی خەو.
٣. تێکەڵی خێزان یان هاوڕێکانت ببە کاتێک تەنیایت.
٤. باوەڕت بە خۆت هەبێت، مێشکت پێویستی بە کاتە بۆ چاکبوونەوە."""

    elif any(word in text for word in arezakan):
        return "ئێستا مۆبایلەکەت دابنێ! هەستە بڕۆ دەرەوە پیاسەیەک بکە یان ئاوێکی سارد بە ڕووخسارتدا بکە. ئەم هەستە کاتییە و تێدەپەڕێت. تۆ بەهێزتری لەوە!"

    elif "سڵاو" in text:
        return "سڵاو! من ڕێبەری تەندروستیی تۆم. دەتوانیت پێم بڵێیت (سەرکەوتم) بۆ ئەوەی ڕۆژێک بۆ ژمێرەرەکەت زیاد بکەم، یان پرسیار لەسەر زیان و چارەسەرەکان بکە."

    else:
        return "تێناگەم مەبەستت چییە. دەتوانیت بڵێیت (سەرکەوتم)، (چەند ڕۆژە)، یان پرسیار دەربارەی (چارەسەر) بکەیت."

print("چاتبۆتی 'ڕێبەری تەندروستی' ئامادەیە!")
print("تایبەتمەندی نوێ: دەتوانیت بڵێیت 'سەرکەوتم' یان 'شکاندم' بۆ هەژمارکردنی ڕۆژەکانت.")
print("-" * 60)

while True:
    user_text = input("بەکارھێنەر: ")

    if "خوات لەگەڵ" in user_text or "بای" in user_text:
        print("ڕێبەری تەندروستی: هیوای ژیانێکی ئارام و تەندروستت بۆ دەخوازم. خوات لەگەڵ!")
        break

    response = recovery_guide_bot(user_text)
    print("ڕێبەری تەندروستی:\n" + response)
    print("-" * 60)
```
